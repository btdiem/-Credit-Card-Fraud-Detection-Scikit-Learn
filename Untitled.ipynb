### Build Machine Learning Model & Validation

```python
X = df.drop(columns=['Class'])
Y = df['Class']
```

```python
print(X.columns)
print(Y.head())
print(X.shape)
print(Y.shape)
```

```python
from sklearn.covariance import EllipticEnvelope
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
```

```python
rng = np.random.RandomState(42)

models = {
     "Isolation Forest" : IsolationForest(max_samples=len(X),
                      random_state=rng, contamination=outlier_fraction)
 #   , "Local Outlier Factor" : LocalOutlierFactor(n_neighbors=20, contamination=outlier_fraction)
#    , "Elliptic Envelope" : EllipticEnvelope(contamination=outlier_fraction)
}
```

```python
for i, (model_name, model) in enumerate(models.items()):
    if model_name == "Local Outlier Factor":
        y_pred = model.fit_predict(X)
#        score_pred = model.negative_outlier_factor_
    else:
        model.fit(X)
#        score_pred = model.decision_function(X)
        y_pred = model.predict(X)
        
    #  Predict the labels (1 inlier, -1 outlier) of X according to the fitted model    
    y_pred[y_pred==1] = 0
    y_pred[y_pred==-1] = 1
    
    n_errors = (y_pred != Y).sum()
    
    print("Model: {} - Errors: {}".format(model_name, n_errors))
    print(accuracy_score(Y, y_pred))
    print(classification_report(Y, y_pred))

    print(y_pred)
# print("model {} with predictve score {} ".format(model_name,y_pred))
        
```
